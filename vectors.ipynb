```python
import numpy as np
```

```python
a = np.array([2, -3, 1])
b = np.array([1, 4, -2])

sum = a + b
sub = a - b
mul = a * b
la = 1 * a + (-1) * b
scl = 3 * a

print(f"Sum: {sum}")
print(f"Sub: {sub}")
print(f"Mul: {mul}")
print(f"LA: {la}")
print(f"Scalar(3): {scl}")
```

```python
# A * x = B
e1 = np.array([1, 0])
e2 = np.array([0, 1])
target = np.array([5, -3])

A = np.column_stack((e1, e2))
print(f"A: {A}")
coeff = np.linalg.solve(A, target)
print(f"Coeff: {coeff}")
```

```python
# v = Î»*u
def is_collinear(u, v):
    u = np.array(u)
    v = np.array(v)

    if u[0] == 0:
        return v[0] == 0

    alfa = v[0] / u[0]
    expected_v1 = alfa * u[1]

    return np.isclose(v[1], expected_v1)


u1 = [2, 6]
v1 = [1, 3]
result1 = is_collinear(u1, v1)
print(f"Vectors {u1} and {v1} collinear? -> {result1}")

u2 = [3, -1]
v2 = [6, 2]
result2 = is_collinear(u2, v2)
print(f"Vectors {u2} and {v2} collinear? -> {result2}")
```
