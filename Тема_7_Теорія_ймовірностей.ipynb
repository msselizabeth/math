### Приклад 1. Геометрична інтерпретація: $P(A) = \cfrac{|A|}{|\Omega|}$

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_euler_diagram():
    # Створюємо фігуру та осі
    fig, ax = plt.subplots(figsize=(6, 6))

    # 1. Малюємо Простір подій Omega (велике коло)
    # (0.5, 0.5) - центр, 0.4 - радіус
    omega_circle = patches.Circle((0.5, 0.5), 0.4,
                                  edgecolor='black',
                                  facecolor='#f0f0f0', # Світло-сірий фон
                                  linewidth=2,
                                  label='Omega')
    ax.add_patch(omega_circle)

    # 2. Малюємо Подію A (менше коло всередині)
    # Зміщуємо трохи центр, щоб виглядало природніше
    event_a_circle = patches.Circle((0.45, 0.45), 0.15,
                                    edgecolor='#2b8cbe', # Синій контур
                                    facecolor='#a6bddb', # Блакитна заливка
                                    alpha=0.8,           # Прозорість
                                    linewidth=2,
                                    label='A')
    ax.add_patch(event_a_circle)

    # 3. Додаємо підписи (LaTeX форматування)
    # Підпис Omega
    ax.text(0.65, 0.65, r'$\Omega$', fontsize=24, color='black', weight='bold')

    # Підпис події A
    ax.text(0.45, 0.45, r'$A$', fontsize=20, color='white',
            ha='center', va='center', weight='bold')

    # Налаштування відображення
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_aspect('equal') # Щоб кола були круглими, а не овальними
    ax.axis('off') # Прибираємо осі координат

    # Зберігаємо або показуємо
    plt.tight_layout()
    plt.show()

draw_euler_diagram()
```

### Приклад 3. Дерево ймовірностей. Витягування 2 куль без повернення

```python
import matplotlib.pyplot as plt


fig, ax = plt.subplots(figsize=(12, 8))

# Налаштування осей
ax.set_xlim(0, 1.2)
ax.set_ylim(0, 1)
ax.axis('off')

# --- КООРДИНАТИ ВУЗЛІВ ---
# Початок для стрілок (припускаємо, що це права межа блока "Старт")
arrow_origin_x = 0.08
arrow_origin_y = 0.5
arrow_start_point = (arrow_origin_x, arrow_origin_y)

# Позиція центру тексту "Старт"
# Якщо права межа блока "Старт" знаходиться на arrow_origin_x,
# і припустити, що півширина блока "Старт" приблизно 0.03 (візуально)
start_text_center_x = arrow_origin_x - 0.03
start_text_center_y = arrow_origin_y
start_text_pos = (start_text_center_x, start_text_center_y)

# Перший рівень (1-ша куля)
# W1 (верх), B1 (низ)
l1_w_pos = (0.35, 0.75)
l1_b_pos = (0.35, 0.25)

# Другий рівень (2-га куля)
# Від W1 йдуть: W2 (верх-верх), B2 (верх-низ)
l2_ww_pos = (0.75, 0.85)
l2_wb_pos = (0.75, 0.65)

# Від B1 йдуть: W2 (низ-верх), B2 (низ-низ)
l2_bw_pos = (0.75, 0.35)
l2_bb_pos = (0.75, 0.15)

# --- ФУНКЦІЯ МАЛЮВАННЯ ГІЛКИ ---
def draw_branch(start, end, prob_text, label_text, color_node):
    # Лінія
    ax.annotate("", xy=end, xytext=start,
                arrowprops=dict(arrowstyle="->", lw=2, color='gray'))

    # Текст ймовірності (посередині лінії)
    mid_x = (start[0] + end[0]) / 2
    mid_y = (start[1] + end[1]) / 2
    ax.text(mid_x - 0.01, mid_y + 0.04, prob_text, fontsize=16,
            ha='center', color='black', weight='bold')

    # Вузол (кружечок)
    circle = plt.Circle(end, 0.04, facecolor=color_node, edgecolor='black', zorder=10)
    ax.add_patch(circle)

    # Підпис вузла (що випало)
    ax.text(end[0], end[1], label_text, fontsize=12,
            ha='center', va='center', color='black' if color_node=='white' else 'white', weight='bold')

# --- МАЛЮЄМО ДЕРЕВО ---

# Корінь
ax.text(start_text_pos[0], start_text_pos[1], "Старт", ha='center', fontsize=12,
        bbox=dict(facecolor='#f0f0f0', edgecolor='black', boxstyle='round'))

# === РІВЕНЬ 1 ===
# Гілка до Білої (W)
draw_branch(arrow_start_point, l1_w_pos, r"$P(Б)=\frac{5}{8}$", "Б", "white")
# Гілка до Чорної (B)
draw_branch(arrow_start_point, l1_b_pos, r"$P(Ч)=\frac{3}{8}$", "Ч", "gray")

# === РІВЕНЬ 2 ===
# Від Білої -> до Білої (Залишилось 4 з 7)
draw_branch(l1_w_pos, l2_ww_pos, r"$P(Б|Б)=\frac{4}{7}$", "Б", "white")

# Від Білої -> до Чорної (Залишилось 3 з 7)
draw_branch(l1_w_pos, l2_wb_pos, r"$P(Ч|Б)=\frac{3}{7}$", "Ч", "gray")

# Від Чорної -> до Білої (Залишилось 5 з 7)
draw_branch(l1_b_pos, l2_bw_pos, r"$P(Б|Ч)=\frac{5}{7}$", "Б", "white")

# Від Чорної -> до Чорної (Залишилось 2 з 7)
draw_branch(l1_b_pos, l2_bb_pos, r"$P(Ч|Ч)=\frac{2}{7}$", "Ч", "gray")


# Форматування тексту результатів
results = [
    (l2_ww_pos, r"$P(Б \cap Б) = \frac{5}{8} \cdot \frac{4}{7} = \frac{20}{56}$", "crimson"),
    (l2_wb_pos, r"$P(Б \cap Ч) = \frac{5}{8} \cdot \frac{3}{7} = \frac{15}{56}$", "black"),
    (l2_bw_pos, r"$P(Ч \cap Б) = \frac{3}{8} \cdot \frac{5}{7} = \frac{15}{56}$", "black"),
    (l2_bb_pos, r"$P(Ч \cap Ч) = \frac{3}{8} \cdot \frac{2}{7} = \frac{6}{56}$", "black"),
]

for pos, text, col in results:
    ax.text(pos[0] + 0.08, pos[1], text, fontsize=14, va='center', color=col)

# Заголовок
plt.tight_layout()
plt.show()
```

### Приклад 4. Програмна реалізація теореми Байєса.

```python
def bayes_theorem(p_h, p_evidence_given_h, p_evidence_given_not_h):
    """
    Обчислює апостеріорну ймовірність гіпотези за теоремою Байєса.

    Параметри:
    p_h: P(H) - апріорна ймовірність гіпотези
    p_evidence_given_h: P(E|H) - ймовірність спостереження за умови що гіпотеза вірна
    p_evidence_given_not_h: P(E|not H) - ймовірність спостереження за умови що гіпотеза хибна

    Повертає:
    P(H|E) - апостеріорна ймовірність гіпотези після спостереження
    """
    # Апріорна ймовірність протилежної гіпотези
    p_not_h = 1 - p_h

    # Повна ймовірність спостереження (evidence)
    p_evidence = p_h * p_evidence_given_h + p_not_h * p_evidence_given_not_h

    # Апостеріорна ймовірність за теоремою Байєса
    p_h_given_evidence = (p_h * p_evidence_given_h) / p_evidence

    return p_h_given_evidence
```

```python
# Приклад 2: Медичний тест
p_disease = 0.01        # P(D) - базова частота хвороби 1%
sensitivity = 0.99      # P(T|D) - чутливість тесту 99%
false_positive = 0.05   # P(T|not D) - хибнопозитивний результат 5%

result = bayes_theorem(p_disease, sensitivity, false_positive)

print(f"Апріорна ймовірність хвороби: {p_disease:.1%}")
print(f"Апостеріорна ймовірність після позитивного тесту: {result:.1%}")
```

```python
# Приклад 1: Спам-фільтр
p_spam = 0.3           # 30% листів це спам
p_word_in_spam = 0.6   # 60% спам-листів містять слово "безкоштовно"
p_word_in_normal = 0.05 # 5% нормальних листів містять це слово

result_spam = bayes_theorem(p_spam, p_word_in_spam, p_word_in_normal)
print(f"Ймовірність спаму при слові 'безкоштовно': {result_spam:.1%}")
```

```python
# Приклад 3: Урни з кулями
p_urn1 = 0.5              # P(H1) - вибрали першу урну
p_white_from_urn1 = 0.3   # P(A|H1) - біла куля з першої урни
p_white_from_urn2 = 0.6   # P(A|H2) - біла куля з другої урни

# Для першої урни (not H1 означає другу урну)
prob_urn1_given_white = bayes_theorem(p_urn1, p_white_from_urn1, p_white_from_urn2)

# Для другої урни
prob_urn2_given_white = bayes_theorem(0.5, p_white_from_urn2, p_white_from_urn1)

print(f"Ймовірність що біла куля з першої урни: {prob_urn1_given_white:.1%}")
print(f"Ймовірність що біла куля з другої урни: {prob_urn2_given_white:.1%}")
```
