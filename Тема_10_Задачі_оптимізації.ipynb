### Приклад 1. Візуалізація для функції однієї змінної

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 3, 500)
f = x**4 - 4*x**3 + 4*x**2

plt.figure(figsize=(10, 6))
plt.plot(x, f, 'b-', linewidth=2, label='$f(x) = x^4 - 4x^3 + 4x^2$')
plt.plot(0, 0, 'ro', markersize=10, label='Локальний мінімум: (0, 0)')
plt.plot(2, 0, 'go', markersize=10, label='Локальний мінімум: (2, 0)')
plt.plot(1, 1, 'mo', markersize=10, label='Локальний максимум: (1, 1)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

### Приклад 2. Візуалізація для функції двох змінних

```python
from mpl_toolkits.mplot3d import Axes3D

x1 = np.linspace(-2, 4, 100)
x2 = np.linspace(-2, 6, 100)
X1, X2 = np.meshgrid(x1, x2)
F = X1**2 + X2**2 - 2*X1 - 4*X2

fig = plt.figure(figsize=(12, 5))

# 3D поверхня
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X1, X2, F, cmap='viridis', alpha=0.8)
ax1.scatter([1], [2], [-5], color='red', s=100, label='Мінімум: (1, 2, -5)')
ax1.set_xlabel('$x_1$')
ax1.set_ylabel('$x_2$')
ax1.set_zlabel('$f(x_1, x_2)$')
ax1.set_title('Поверхня функції')

# Контурні лінії
ax2 = fig.add_subplot(122)
contour = ax2.contour(X1, X2, F, levels=20, cmap='viridis')
ax2.plot(1, 2, 'ro', markersize=10, label='Мінімум: (1, 2)')
ax2.set_xlabel('$x_1$')
ax2.set_ylabel('$x_2$')
ax2.set_title('Контурні лінії')
ax2.legend()
plt.colorbar(contour, ax=ax2)

plt.tight_layout()
plt.show()
```

### Приклад 4. Функція $f(x) = x^3$ має стаціонарну точку в $x = 0$, де $f'(0) = 0$, але це точка перегину, а не екстремум

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Налаштування даних
x = np.linspace(-1.5, 1.5, 400)
y = x**3

# 2. Створення графіка
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='$f(x) = x^3$', color='blue', linewidth=2)

# 3. Додавання дотичної (y = 0)
# Відображаємо, що в точці x=0 похідна (нахил) дорівнює 0
plt.plot(x, np.zeros_like(x), '--', color='gray', alpha=0.7, label="Дотична (f'(0) = 0)")

# 4. Виділення точки перегину
plt.scatter([0], [0], color='red', s=100, zorder=5, label='Точка перегину (0, 0)')

# 5. Налаштування осей "як у підручнику" (через центр)
ax = plt.gca()
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# 6. Оформлення та підписи
plt.legend(loc='upper left')
plt.grid(True, alpha=0.3)

plt.show()
```

### Приклад 7. Мінімізація квадратичної функції

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar

# Визначаємо функцію
def f(x):
    return (x - 3)**2 + 2

# Знаходимо мінімум
result = minimize_scalar(f, bounds=(0, 6), method='bounded')

print("Результати оптимізації:")
print(f"Мінімум знайдено в точці: x = {result.x:.6f}")
print(f"Значення функції: f(x) = {result.fun:.6f}")
print(f"Успішність: {result.success}")
print(f"Кількість ітерацій: {result.nit}")
print(f"Повідомлення: {result.message}")
```

```python
# Створюємо точки для графіка
x = np.linspace(0, 6, 500)
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2, label='$f(x) = (x-3)^2 + 2$')
plt.plot(result.x, result.fun, 'ro', markersize=12,
         label=f'Мінімум: ({result.x:.2f}, {result.fun:.2f})')
plt.axvline(x=result.x, color='r', linestyle='--', alpha=0.3)
plt.axhline(y=result.fun, color='r', linestyle='--', alpha=0.3)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)
plt.show()
```

### Приклад 8. Функція з кількома локальними мінімумами

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar

def f_complex(x):
    return np.sin(x) + 0.1*x

# Знаходимо мінімум
result = minimize_scalar(f_complex, bounds=(0, 15), method='bounded')

print("Результати оптимізації:")
print(f"Мінімум знайдено в точці: x = {result.x:.6f}")
print(f"Значення функції: f(x) = {result.fun:.6f}")
```

### Приклад 9. Оптимізація функції двох змінних

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Визначаємо функцію двох змінних
def f_2d(x):
    return x[0]**2 + 2*x[1]**2 - 2*x[0] - 8*x[1] + 10

# Початкова точка пошуку
x0 = np.array([0.0, 0.0])

# Знаходимо мінімум методом BFGS
result = minimize(f_2d, x0, method='BFGS')

print("Результати оптимізації:")
print(f"Мінімум знайдено в точці: x1 = {result.x[0]:.6f}, x2 = {result.x[1]:.6f}")
print(f"Значення функції: f(x) = {result.fun:.6f}")
print(f"Успішність: {result.success}")
print(f"Кількість ітерацій: {result.nit}")
print(f"Кількість обчислень функції: {result.nfev}")
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Створюємо сітку для візуалізації
x1 = np.linspace(-2, 4, 100)
x2 = np.linspace(-1, 5, 100)
X1, X2 = np.meshgrid(x1, x2)
Z = X1**2 + 2*X2**2 - 2*X1 - 8*X2 + 10

fig = plt.figure(figsize=(14, 6))

# 3D поверхня
ax1 = fig.add_subplot(121, projection='3d')
surf = ax1.plot_surface(X1, X2, Z, cmap='viridis', alpha=0.8)
ax1.scatter([result.x[0]], [result.x[1]], [result.fun],
            color='red', s=100, marker='o', label='Мінімум')
ax1.set_xlabel('$x_1$', fontsize=11)
ax1.set_ylabel('$x_2$', fontsize=11)
ax1.set_zlabel('$f(x_1, x_2)$', fontsize=11)
ax1.set_title('Поверхня функції', fontsize=13)
fig.colorbar(surf, ax=ax1, shrink=0.5)

# Контурні лінії
ax2 = fig.add_subplot(122)
contour = ax2.contour(X1, X2, Z, levels=20, cmap='viridis')
ax2.plot(result.x[0], result.x[1], 'ro', markersize=12,
         label=f'Мінімум: ({result.x[0]:.1f}, {result.x[1]:.1f})')
ax2.plot(x0[0], x0[1], 'bs', markersize=10,
         label=f'Початкова точка: ({x0[0]:.0f}, {x0[1]:.0f})')
ax2.set_xlabel('$x_1$', fontsize=11)
ax2.set_ylabel('$x_2$', fontsize=11)
ax2.set_title('Контурні лінії', fontsize=13)
ax2.legend(fontsize=10)
ax2.grid(True, alpha=0.3)
plt.colorbar(contour, ax=ax2)

plt.tight_layout()
plt.show()
```

### Приклад 10. Порівняння різних методів оптимізації

```python
import numpy as np
from scipy.optimize import minimize

def rosenbrock(x):
    return (1 - x[0])**2 + 100*(x[1] - x[0]**2)**2

# Початкова точка
x0 = np.array([0.0, 0.0])

# Методи для порівняння
methods = ['BFGS', 'Nelder-Mead', 'CG', 'Powell']

print("Порівняння методів оптимізації функції Розенброка:\n")
print(f"{'Метод':<15} {'x1':>10} {'x2':>10} {'f(x)':>12} {'Ітерації':>10} {'Обчислень':>11}")
print("-" * 75)

for method in methods:
    result = minimize(rosenbrock, x0, method=method)
    print(f"{method:<15} {result.x[0]:>10.6f} {result.x[1]:>10.6f} "
          f"{result.fun:>12.10f} {result.nit:>10} {result.nfev:>11}")
```

```python
x1 = np.linspace(-2, 2, 400)
x2 = np.linspace(-1, 3, 400)
X1, X2 = np.meshgrid(x1, x2)
Z = (1 - X1)**2 + 100*(X2 - X1**2)**2

fig = plt.figure(figsize=(14, 6))

# 3D поверхня (нормальна шкала)
ax1 = fig.add_subplot(121, projection='3d')
surf = ax1.plot_surface(X1, X2, Z, cmap='viridis', alpha=0.8)
ax1.scatter([1], [1], [0], color='red', s=100, marker='o')
ax1.set_xlabel('$x_1$', fontsize=11)
ax1.set_ylabel('$x_2$', fontsize=11)
ax1.set_zlabel('$f(x_1, x_2)$', fontsize=11)
ax1.set_title('Функція Розенброка', fontsize=13)
fig.colorbar(surf, ax=ax1, shrink=0.5)

# Контурні лінії (тепер на нормальній шкалі)
ax2 = fig.add_subplot(122)
levels = np.linspace(Z.min(), Z.max(), 20) # Змінено на лінійну шкалу
contour = ax2.contour(X1, X2, Z, levels=levels, cmap='viridis')
ax2.plot(1, 1, 'ro', markersize=12, label='Глобальний мінімум: (1, 1)')
ax2.plot(0, 0, 'bs', markersize=10, label='Початкова точка: (0, 0)')
ax2.set_xlabel('$x_1$', fontsize=11)
ax2.set_ylabel('$x_2$', fontsize=11)
ax2.set_title('Контурні лінії функції Розенброка', fontsize=13)
ax2.legend(fontsize=10)
ax2.grid(True, alpha=0.3)
plt.colorbar(contour, ax=ax2)

plt.tight_layout()
plt.show()
```

### Приклад 11. Задачу мінімізації цільової функції

```python
from scipy.optimize import linprog

# Мінімізуємо: -3x - 2y
c = [-3, -2]

# Обмеження: Ax <= b
A = [
    [1, 2],
    [3, 2]
]
b = [8, 12]

# Обмеження на змінні: x ≥ 0, y ≥ 0
bounds = [(0, None), (0, None)]

res = linprog(c, A_ub=A, b_ub=b, bounds=bounds, method='highs')

print("Статус:", res.message)
print("Оптимальне значення:", -res.fun)
print("Оптимальне розв'язання:", res.x)
```
