# Task 1

Компанія продає бездротові навушники і хоче встановити ціну, яка максимізує прибуток.

За результатами досліджень, попит на навушники залежить від ціни за формулою:

$Q(x)=2000−0.8x$
- $x$ — ціна в гривнях, 
- $Q(x)$ — кількість проданих одиниць на місяць.

Собівартість одних навушників становить 800 грн. Тоді прибуток компанії:

$P(x)=(x−800)⋅Q(x)=(x−800)(2000−0.8x)$

Завдання:

- Знайдіть критичну точку аналітично. Розкрийте дужки у функції прибутку $P(x)$ та спростіть. Обчисліть похідну $P′(x)$. Прирівняйте похідну до нуля та розв'яжіть рівняння $P′(x)=0$, щоб знайти критичну точку $x^*$.

    $P(x) = 2000x - 0.8x^2 - 1600000 + 640x = -0.8x^2 + 2640x - 1600000$

    $P′(x) = -1.6x + 2640$

    $-1.6x + 2640 = 0$

    $-1.6x = -2640$
    
    $x = 1650$

- Перевірте тип екстремуму.  Аналітично обчисліть другу похідну $P′′(x)$. Визначте її знак та зробіть висновок: чи є знайдена точка максимумом чи мінімумом.
    
    $P′′(x) = -1.6 $ $=>$ знайдена точка є максимумом при ціні 1650грн.

- Порівняйте результати та обчисліть прибуток. Порівняйте аналітичний та чисельний результати. Аналітично обчисліть максимальний місячний прибуток при оптимальній ціні.

    $P(x) = (1650 - 800)(2000 - 0.8 \cdot 1650) = 578000$

    *Висновок:* Аналітичний та програмний результати співпали. Різниця $0.0000000001$ — це стандартна похибка обчислень з плаваючою комою.

```python
from scipy.optimize import minimize_scalar

# Negative P(x)
def p(x):
    return - (x - 800) *(2000 - 0.8 * x)

# Find a min
result = minimize_scalar(p, bounds=(800, 2500), method='bounded')
print(f"Result: {result.message}")
print(f"Min x: {result.x}")

# Check profit
profit = p(result.x)
print(f"Profit: {profit * -1}") # multiply by -1 to return the original func
```

# Task 2

Розглянемо функцію:

$f(x,y)=x^2+xy+y ^2−6x−9y+20$

Завдання

- Знайдіть критичну точку аналітично. Обчисліть частинні похідні $\frac{∂f}{∂x}$ та $\frac{∂f}{∂y}$ ю Прирівняйте їх до нуля та розв'яжіть систему рівнянь, щоб знайти критичну точку ($x^*$, $y^*$)

    $\frac{∂f}{∂x} = 2x + y - 6$
    
    $\frac{∂f}{∂y} = x + 2y - 9$

    $\begin{cases} 2x + y - 6 = 0 \\ x + 2y - 9 = 0 \end{cases}$
    $\begin{cases} 2x + y - 6 = 0 \\ y = \frac{9 - x}{2} \end{cases}$
    $\begin{cases} 2x + \frac{9 - x}{2} - 6 = 0 \\ y = \frac{9 - x}{2} \end{cases}$
    $\begin{cases} \frac{4x + 9 - x}{2} = 6 \\ y = \frac{9 - x}{2} \end{cases}$
    $\begin{cases} 3x + 9 = 12 \\ y = \frac{9 - x}{2} \end{cases}$
    $\begin{cases} 3x = 3  \\ y = \frac{9 - x}{2} \end{cases}$
    $\begin{cases} x = 1 \\ y = \frac{9 - 1}{2} \end{cases}$
    $\begin{cases} x = 1 \\ y = 4 \end{cases}$

  ($x^*$, $y^*$) = $(1,4)$

- Перевірте тип екстремуму. Далі аналітично обчисліть другі частинні похідні та складіть матрицю Гессе. Знайдіть власні значення матриці через характеристичне рівняння $det(H−λI)=0$. 

  $\frac{∂^2f}{∂x^2} = 2$ ;  $\frac{∂^2f}{∂y^2} = 2$;   $\frac{∂^2f}{∂x∂y}= 1$ ;  $\frac{∂^2f}{∂y∂x} = 1$ 

  $H = \begin{pmatrix} 2 & 1 \\ 1 & 2 \end{pmatrix}$
  $\lambda_{1} = 1$ $\lambda_{2} = 3$

  $\lambda_{1}, \lambda_{2} > 0 => $ (1,4) це мінімум.



```python
import numpy as np
from scipy.optimize import minimize


# f(x)
def f(x):
    return x[0] ** 2 + x[0] * x[1] + x[1] ** 2 - 6 * x[0] - 9 * x[1] + 20


# Start points
x0 = np.array([0.0, 0.0])
x1 = np.array([10.0, 10.0])
x2 = np.array([-5.0, 15.0])


# Find a min using BFGS
res_0 = minimize(f, x0, method="BFGS")

# Check stability
res_1 = minimize(f, x1, method="BFGS")
res_2 = minimize(f, x2, method="BFGS")

print(f"f(x) = {res_0.fun:.6f}")
print(f"Мin(start at (0,0)): x = {res_0.x[0]:.6f}, y = {res_0.x[1]:.6f}")
print(f"Мin(start at (10,10)): x = {res_1.x[0]:.6f}, y = {res_1.x[1]:.6f}")
print(f"Мin(start at (-5,15)): x = {res_2.x[0]:.6f}, y = {res_2.x[1]:.6f}")
```
