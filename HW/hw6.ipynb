# Task 1

Компанія продає бездротові навушники і хоче встановити ціну, яка максимізує прибуток.

За результатами досліджень, попит на навушники залежить від ціни за формулою:

$Q(x)=2000−0.8x$
- $x$ — ціна в гривнях, 
- $Q(x)$ — кількість проданих одиниць на місяць.

Собівартість одних навушників становить 800 грн. Тоді прибуток компанії:

$P(x)=(x−800)⋅Q(x)=(x−800)(2000−0.8x)$

Завдання:

- Знайдіть критичну точку аналітично. Розкрийте дужки у функції прибутку $P(x)$ та спростіть. Обчисліть похідну $P′(x)$. Прирівняйте похідну до нуля та розв'яжіть рівняння $P′(x)=0$, щоб знайти критичну точку $x^*$.

    $P(x) = 2000x - 0.8x^2 - 1600000 + 640x = -0.8x^2 + 2640x - 1600000$

    $P′(x) = -1.6x + 2640$

    $-1.6x + 2640 = 0$

    $-1.6x = -2640$
    
    $x = 1650$

- Перевірте тип екстремуму.  Аналітично обчисліть другу похідну $P′′(x)$. Визначте її знак та зробіть висновок: чи є знайдена точка максимумом чи мінімумом.
    
    $P′′(x) = -1.6 $ $=>$ знайдена точка є максимумом при ціні 1650грн.

- Порівняйте результати та обчисліть прибуток. Порівняйте аналітичний та чисельний результати. Аналітично обчисліть максимальний місячний прибуток при оптимальній ціні.

    $P(x) = (1650 - 800)(2000 - 0.8 \cdot 1650) = 578000$

    *Висновок:* Аналітичний та програмний результати співпали. Різниця $0.0000000001$ — це стандартна похибка обчислень з плаваючою комою.

```python
from scipy.optimize import minimize_scalar

# Negative P(x)
def p(x):
    return - (x - 800) *(2000 - 0.8 * x)

# Find a min
result = minimize_scalar(p, bounds=(800, 2500), method='bounded')
print(f"Result: {result.message}")
print(f"Min x: {result.x}")

# Check profit
profit = -p(result.x) # -1 to return the original func
print(f"Profit: {profit}")
```

# Task 2

Розглянемо функцію:

$f(x,y)=x^2+xy+y ^2−6x−9y+20$

Завдання

- Знайдіть критичну точку аналітично. Обчисліть частинні похідні $\frac{∂f}{∂x}$ та $\frac{∂f}{∂y}$ ю Прирівняйте їх до нуля та розв'яжіть систему рівнянь, щоб знайти критичну точку ($x^*$, $y^*$)

    $\frac{∂f}{∂x} = 2x + y - 6$
    
    $\frac{∂f}{∂y} = x + 2y - 9$

    $\begin{cases} 2x + y - 6 = 0 \\ x + 2y - 9 = 0 \end{cases}$
    $\begin{cases} 2x + y - 6 = 0 \\ y = \frac{9 - x}{2} \end{cases}$
    $\begin{cases} 2x + \frac{9 - x}{2} - 6 = 0 \\ y = \frac{9 - x}{2} \end{cases}$
    $\begin{cases} \frac{4x + 9 - x}{2} = 6 \\ y = \frac{9 - x}{2} \end{cases}$
    $\begin{cases} 3x + 9 = 12 \\ y = \frac{9 - x}{2} \end{cases}$
    $\begin{cases} 3x = 3  \\ y = \frac{9 - x}{2} \end{cases}$
    $\begin{cases} x = 1 \\ y = \frac{9 - 1}{2} \end{cases}$
    $\begin{cases} x = 1 \\ y = 4 \end{cases}$

  ($x^*$, $y^*$) = $(1,4)$

- Перевірте тип екстремуму. Далі аналітично обчисліть другі частинні похідні та складіть матрицю Гессе. Знайдіть власні значення матриці через характеристичне рівняння $det(H−λI)=0$. 

  $\frac{∂^2f}{∂x^2} = 2$ ;  $\frac{∂^2f}{∂y^2} = 2$;   $\frac{∂^2f}{∂x∂y}= 1$ ;  $\frac{∂^2f}{∂y∂x} = 1$ 

  $H = \begin{pmatrix} 2 & 1 \\ 1 & 2 \end{pmatrix}$
  $\lambda_{1} = 1$ $\lambda_{2} = 3$

  $\lambda_{1}, \lambda_{2} > 0 => $ (1,4) це мінімум.



```python
import numpy as np
from scipy.optimize import minimize


# f(x)
def f(x):
    return x[0] ** 2 + x[0] * x[1] + x[1] ** 2 - 6 * x[0] - 9 * x[1] + 20


# Start points
x0 = np.array([0.0, 0.0])
x1 = np.array([10.0, 10.0])
x2 = np.array([-5.0, 15.0])


# Find a min using BFGS
res_0 = minimize(f, x0, method="BFGS")

# Check stability
res_1 = minimize(f, x1, method="BFGS")
res_2 = minimize(f, x2, method="BFGS")

print(f"f(x) = {res_0.fun:.6f}")
print(f"Мin(start at (0,0)): x = {res_0.x[0]:.6f}, y = {res_0.x[1]:.6f}")
print(f"Мin(start at (10,10)): x = {res_1.x[0]:.6f}, y = {res_1.x[1]:.6f}")
print(f"Мin(start at (-5,15)): x = {res_2.x[0]:.6f}, y = {res_2.x[1]:.6f}")
```

# Task 3

Меблева фабрика виготовляє два види продукції: стільці та столи. Для виробництва використовуються два ресурси: деревина та робочий час.

| Ресурс            | Витрати на 1 стілець | Витрати на 1 стіл | Загальний запас |
|-------------------|----------------------|-------------------|-----------------|
| Деревина (м²)     | 2                    | 4                 | 120             |
| Робочий час (год) | 3                    | 2                 | 90              |


Прибуток від продажу одного стільця — 500 грн, одного столу — 800 грн.

Потрібно знайти оптимальний план виробництва, який максимізує прибуток.


Завдання:

- Сформулюйте задачу лінійного програмування аналітично. Введіть змінні. 

    $P(x,y) = 500x + 800y$

    Constraints:
    $\begin{cases} 2x + 4y <= 120 \\ 3x + 2y <= 90 \\ x >= 0 \\ y >= 0 \end{cases}$

***Висновок:***

*Математичний розрахунок показав оптимальну кількість у 22.5 стола. Оскільки ми не можемо виготовити половину виробу, доцільно затвердити виробництво на 23 столи та 14 стільців. При цьому на складі не залишиться невикористана деревина, а час буде в залишку.*

- **Аналітичні розраxунки з урухуванням цього:**

    $Прибуток = 500 * 14 + 800 * 23 = 25400$

    Ресурси:
    - $Деревина = 2 * 14 + 4 * 23 = 120$
    - $Час = 3 * 14 + 2 * 23 = 88$

    Залишки:
    - $Деревина = 120 - 120 = 0$
    - $Час = 90 - 88 = 2$

```python
from scipy.optimize import linprog

# Minimize: -500x - 800y
c = [-500, -800]

# Constraints: Ax <= b
A = [
    [2, 4],
    [3, 2]
]
b = [120, 90]

# x ≥ 0, y ≥ 0
bounds = [(0, None), (0, None)]

res = linprog(c, A_ub=A, b_ub=b, bounds=bounds, method='highs')

print("Status:", res.message)
print("Optimal:", -res.fun)
print("Solution:", res.x)

#  Results
wood = 2 * res.x[0] + 4 * res.x[1]
print(f"Optimal wood usage: {wood}. Left: {120 - wood}")

time = 3 * res.x[0] + 2 * res.x[1]
print(f"Optimal time usage: {time}. Left: {90 - time}")
```

# Task 4

Продуктивність (одиниць на годину) змінюється протягом дня за формулою:

$P(t)=100+40t−4t $

- $t$ — час у годинах від початку зміни $0≤t≤10$.

Вартість виробництва однієї одиниці залежить від двох параметрів: витрат на сировину $x$ та витрат на енергію $y$ (грн):

$ C(x,y)=x^2 +y^2−10x−8y+50$

Початкові значення параметрів визначаються з бюджетних обмежень:

$\begin{cases} 2x+y=20 \\ x+3y=25 \end{cases}$
​



```python
import numpy as np
from scipy.optimize import approx_fprime, minimize_scalar, minimize
from scipy.integrate import quad

# Productivity Analysis 
def P(t):
    return 100 + 40*t - 4*t**2

def P_wrapper(t_array):
    return P(t_array[0])

points = [2, 5, 8]
epsilon = 1e-6

for t_val in points:
    derivative = approx_fprime([t_val], P_wrapper, epsilon)[0]
    
    direction = "increasing" if derivative > 0 else "decreasing"
    print(f"Time t={t_val}: P'(t) ≈ {derivative:.2f} ({direction})")

#  Peak productivity
def negative_P(t):
    return -P(t)

res_max = minimize_scalar(negative_P, bounds=(0, 10), method='bounded')
t_peak = res_max.x
P_peak = P(t_peak)
print(f"Peak t = {t_peak:.4f} hours")
print(f"Max productivity: P(t*) = {P_peak:.4f} units/hour")

total_production, error = quad(P, 0, 10)
print(f"Total prod: {total_production:.2f}")
print(f"Error: {error}")

# Cost Optimization 

#  Define matrix A and vector b
A = [[2,1],[1,3]]
b = [20,25]

# Solving the system
initial_params = np.linalg.solve(A, b)
x0, y0 = initial_params
print(f"Initial: x0 = {x0:.2f}, y0 = {y0:.2f}")

# С(x,y)
def cost_function(vars):
    x, y = vars
    return x**2 + y**2 - 10*x - 8*y + 50

# Minimize using BFGS
res = minimize(cost_function, initial_params, method='BFGS')

print(f"Optimal params: x* = {res.x[0]:.4f}, y* = {res.x[1]:.4f}")
print(f"Minimum unit cost = {res.fun:.2f} UAH")

# Total cost
total_cost = total_production * res.fun
print(f"Total productovity cost = {total_cost:.4f} UAH")
```
