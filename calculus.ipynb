# Limits

```python
import sympy as sp

x = sp.symbols('x') # x is math symbol

# f(x) = sin(x) / x
f = sp.sin(x) / x
# f(x) x -> 0
limit = sp.limit(f, x, 0)

# g(x) = (x**2- 1) / (x -1)
g = (x**2 - 1) / (x - 1)
# g(x) when x -> 1
limit_g = sp.limit(g, x, 1)

print(f"For the fuction: {f} the limit is {limit}")
print(f"For the fuction: {g} the limit is {limit_g}")
```

# Deriative

```python
import sympy as sp

x = sp.symbols("x")  # x is math symbol
f = x**3 - 6 * x**2 + 9 * x + 1
g = x * sp.log(x)

deriative = sp.diff(f, x)
deriative_g = sp.diff(g, x)


print(f"Deriative f(x): {deriative}")
print(f"Deriative g(x): {deriative_g}")

slope = deriative_g.subs(x, 1)
print(f"Slope g(x) when x = 1 : {slope}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def L_w(w):
    return w**2 - 6*w + 14

def derivative_L(w):
    return 2 * w - 6

# 1.   (  )
w_vals = np.linspace(-1, 7, 100) 
l_vals = L_w(w_vals)
l_der_vals = derivative_L(w_vals) # deriative values for y-axis

# 2.   (  )
learning_rate = 0.1  #  0.5,  
w0 = 0

# ,     (  )
path_w = [w0]
path_l = [L_w(w0)]

print(f"{'':<5} | {'w ()':<10} | {'Loss ()'}")
print("-" * 35)

#  
for i in range(10):
    grad = derivative_L(w0)
    w_new = w0 - learning_rate * grad
    
    w0 = w_new
    
    #  
    path_w.append(w0)
    path_l.append(L_w(w0))
    
    print(f"{i+1:<5} | {w0:<10.4f} | {L_w(w0):.4f}")

# 
plt.figure(figsize=(10, 6))

#    ()
plt.plot(w_vals, l_vals, label='  $L(w)$', color='blue', alpha=0.6)

# 
plt.plot(w_vals, l_der_vals, label=' $L\'(w) = 2w - 6$', linestyle='--')

#    (  + )
plt.plot(path_w, path_l, color='red', marker='o', linestyle='--', label='  ')

# 
plt.title(f"  (Learning Rate = {learning_rate})")
plt.xlabel(' w')
plt.ylabel(' L(w)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1.   ( )
# X -  (), Y -  
X = np.array([1, 2, 3, 4])
Y = np.array([2, 4, 6, 8])  #  ,     y = 2x

# 2. 
# w -  "" ().    ( )
w = 0.0 

#  
def predict(x, weight):
    return x * weight

# 3.  
learning_rate = 0.01
epochs = 28 #    

plt.figure(figsize=(10, 6))

#    ( )
plt.scatter(X, Y, color='blue', s=100, label='  (Truth)')

print(f"{'Epoch':<5} | {'w':<10} | {'Error (MSE)':<10} | {'Gradient'}")
print("-" * 45)

# 4.   ( )
for epoch in range(epochs):
    # .  (Forward pass)
    Y_pred = predict(X, w)
    
    # .   (Loss) -    
    loss = np.mean((Y_pred - Y)**2)
    
    # .   ( !)
    # : dL/dw = 2 * ( - ) * 
    # mean() ,     4 ,   
    gradient = np.mean(2 * (Y_pred - Y) * X)
    
    # .   ( )
    w = w - learning_rate * gradient
    
    print(f"{epoch+1:<5} | {w:<10.4f} | {loss:<10.4f} | {gradient:.4f}")

    #    (  )
    if epoch % 4 == 0: #   4 
        plt.plot(X, predict(X, w), linestyle='--', alpha=0.5, label=f' {epoch} (w={w:.2f})')

#   
plt.plot(X, predict(X, w), color='red', linewidth=3, label=f' (w={w:.2f})')

plt.title("     ")
plt.xlabel(" X")
plt.ylabel(" Y")
plt.legend()
plt.grid(True)
plt.show()
```
