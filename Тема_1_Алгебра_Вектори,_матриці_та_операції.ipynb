### Приклад 1. Геометричне представлення трьох векторів на площині.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Створюємо декілька векторів
vectors = {
    'v1': np.array([3, 2]),
    'v2': np.array([-2, 3]),
    'v3': np.array([1, -2])
}

# Візуалізація
fig, ax = plt.subplots(figsize=(8, 8))

colors = ['blue', 'red', 'green']
for (name, vec), color in zip(vectors.items(), colors):
    ax.arrow(0, 0, vec[0], vec[1],
             head_width=0.3, head_length=0.3,
             fc=color, ec=color, linewidth=2, label=name)
    # Підписуємо координати
    ax.text(vec[0], vec[1] + 0.4, f'{name} = [{vec[0]}, {vec[1]}]',
            fontsize=10, ha='center')

ax.grid(True, alpha=0.3)
ax.set_xlim(-4, 4)
ax.set_ylim(-3, 4)
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
ax.set_xlabel('$v_1$', fontsize=12)
ax.set_ylabel('$v_2$', fontsize=12)
ax.set_title('Геометричне представлення векторів на площині', fontsize=14)
ax.legend()
ax.set_aspect('equal')
plt.show()
```

### Приклад 4. Геометрична візуалізація додавання векторів

```python
# import matplotlib.pyplot as plt
# import numpy as np

v = np.array([1, 2])
w = np.array([3, -1])
result = v + w

fig, ax = plt.subplots(figsize=(8, 6))

# Перший вектор v (синій)
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1,
          color='blue', width=0.006)
ax.text(0.3, 1.3, r'$\mathbf{a}$', fontsize=14, color='blue')

# Другий вектор w прикладений до кінця v (червоний)
ax.quiver(v[0], v[1], w[0], w[1], angles='xy', scale_units='xy', scale=1,
          color='red', width=0.006)
ax.text(2.8, 1.8, r'$\mathbf{b}$', fontsize=14, color='red')

# Результуючий вектор v+w (зелений, товстіший)
ax.quiver(0, 0, result[0], result[1], angles='xy', scale_units='xy', scale=1,
          color='green', width=0.008, linewidth=2)
ax.text(2.2, 0.3, r'$\mathbf{a}+\mathbf{b}$', fontsize=14, color='green')

# Точки для наочності
ax.plot([0, v[0], result[0]], [0, v[1], result[1]], 'ko', markersize=4)

ax.set_xlim(-0.5, 5)
ax.set_ylim(-0.5, 3)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Додавання векторів')

plt.show()
```

### Приклад 6. Множення вектора на скаляр

```python
# Базовий вектор
v = np.array([2, 3])

# Множення на різні скаляри
scalars = [0.5, 1, 2, -1]
results = {}

print("Множення вектора v = [2, 3] на різні скаляри:\n")
for alpha in scalars:
    result = alpha * v
    results[alpha] = result
    print(f"  {alpha} × v = {result}")
```

### Приклад 9. Лінійної комбінації векторів

```python
# Одиничні вектори вздовж осей координат
e1 = np.array([1, 0])  # одиничний вектор по осі x
e2 = np.array([0, 1])  # одиничний вектор по осі y

# Довільний вектор
target = np.array([3, 2])

# Представлення через одиничні вектори
result = 3*e1 + 2*e2

print("Одиничні вектори вздовж осей:")
print(f"  e₁ = {e1} (вздовж осі x)")
print(f"  e₂ = {e2} (вздовж осі y)")
print(f"\nЦільовий вектор: {target}")
print(f"\nЛінійна комбінація:")
print(f"  3×e₁ + 2×e₂ = 3×{e1} + 2×{e2} = {result}")
print(f"\nПеревірка: {np.array_equal(result, target)}")
```

### Приклад 11.  Візуалізація кольорів RGB

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Основні кольори у RGB-просторі (значення від 0 до 255)
red = np.array([255, 0, 0])
green = np.array([0, 255, 0])
blue = np.array([0, 0, 255])

# Отримуємо нові кольори через лінійні комбінації
yellow = 0.5*red + 0.5*green
purple = 0.5*red + 0.5*blue
cyan = 0*red + 0.5*green + 0.5*blue
white = 0.33*red + 0.33*green + 0.33*blue

# Створюємо словник
colors = {
    "Червоний": red,
    "Зелений": green,
    "Синій": blue,
    "Оливковий": yellow.astype(int),
    "Пурпуровий": purple.astype(int),
    "Темно-бірюзовий": cyan.astype(int),
    "Темно-сірий": white.astype(int)
}

# Створюємо фігуру та осі для малювання
fig, ax = plt.subplots(figsize=(6, 4), dpi=100)

# Проходимо по кожному кольору і додаємо його на зображення
for i, (name, color_rgb) in enumerate(colors.items()):
    # Matplotlib очікує кольори у форматі RGB з значеннями від 0 до 1,
    # тому нормалізуємо значення, поділивши на 255
    normalized_color = color_rgb / 255.0

    # Створюємо прямокутник відповідного кольору
    rect = patches.Rectangle(
        (0.1, 0.85 - i * 0.12), # координати (x, y) лівого нижнього кута
        0.25,                  # ширина
        0.1,                   # висота
        facecolor=normalized_color,
        edgecolor='black',     # колір рамки
        linewidth=1
    )

    # Додаємо прямокутник на осі
    ax.add_patch(rect)

    # Додаємо текстовий підпис з назвою та RGB-значенням
    ax.text(
        0.4,                           # координата x для тексту
        0.9 - i * 0.12,                # координата y для тексту
        f"{name}: {color_rgb}",         # сам текст
        fontsize=12,
        verticalalignment='center'     # вирівнювання по центру
    )

# Налаштування вигляду графіка
ax.set_xlim(0, 1.2)
ax.set_ylim(0, 1)
ax.axis('off') # Вимикаємо осі координат
ax.set_title("Візуалізація кольорів RGB", fontsize=14)

# Показуємо зображення
plt.show()
```

### Приклад 20. Візуалізація механізму транспонування

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Матриця для візуалізації
A = np.array([[1, 2],
              [3, 4],
              [5, 6]])

A_T = A.T

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

def draw_matrix(ax, matrix, title):
    ax.set_xlim(-0.5, matrix.shape[1] + 0.5)
    ax.set_ylim(-0.5, matrix.shape[0] + 0.5)
    ax.set_aspect('equal')
    ax.set_title(title, fontsize=16, fontweight='bold')
    ax.axis('off')

    rows, cols = matrix.shape

    for i in range(rows):
        for j in range(cols):
            # Highlight diagonal elements for square matrices
            if i == j: # Check for main diagonal
                color = 'lightcoral' # Highlight diagonal
            else:
                color = 'lightblue'

            rect = Rectangle((j, rows-1-i), 1, 1,
                            linewidth=2.5, edgecolor='black',
                            facecolor=color, alpha=0.8)
            ax.add_patch(rect)

            ax.text(j + 0.5, rows-1-i + 0.5, str(matrix[i, j]),
                   ha='center', va='center', fontsize=18, fontweight='bold')

    # Add row and column labels
    for i in range(rows):
        ax.text(-0.3, rows-1-i + 0.5, f'i={i+1}',
               ha='right', va='center', fontsize=10, style='italic')
    for j in range(cols):
        ax.text(j + 0.5, rows + 0.2, f'j={j+1}',
               ha='center', va='bottom', fontsize=10, style='italic')


draw_matrix(ax1, A, 'A (3×2)')
draw_matrix(ax2, A.T, 'Aᵀ (2×3)')

fig.text(0.45, 0.5, '→', fontsize=40, ha='center', va='center')
fig.text(0.45, 0.4, 'транспонування', fontsize=12, ha='center', va='center', style='italic')

plt.tight_layout()
plt.show()
```

### Приклад 22. Транспонування в контексті datasets

```python
# Dataset: 3 студенти, 4 предмети
grades = np.array([
    [85, 90, 78, 92],  # студент 0
    [88, 85, 91, 87],  # студент 1
    [92, 89, 85, 90]   # студент 2
])

print("Оригінальний dataset (студенти × предмети):")
print(grades)
print(f"Розмірність: {grades.shape}")

# Транспонуємо: тепер рядки — предмети, стовпці — студенти
grades_T = grades.T

print("\nТранспонований dataset (предмети × студенти):")
print(grades_T)
print(f"Розмірність: {grades_T.shape}")

# Тепер легко обчислити середній бал по кожному предмету
print("\nСередній бал по предметах:")
for i, avg in enumerate(grades_T.mean(axis=1)):
    print(f"  Предмет {i+1}: {avg:.1f}")
```

### Приклад 24. Операції над матрицями

```python
import numpy as np

# Дві матриці 2×3
A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([[7, 8, 9],
              [10, 11, 12]])

print("Матриця A:")
print(A)
print("\nМатриця B:")
print(B)

# Додавання
C = A + B
print("\nA + B:")
print(C)

# Віднімання
D = A - B
print("\nA - B:")
print(D)

# Множення на скаляр
E = 3 * A
print("\n3 × A:")
print(E)

# Комбінована операція
F = 2*A + 3*B
print("\n2×A + 3×B:")
print(F)
```

### Приклад 25. Нормалізація датасету — віднімання середнього та ділення на стандартне відхилення

```python
# Dataset: 5 зразків, 3 ознаки
data = np.array([[170, 70, 25],
                 [165, 60, 30],
                 [180, 80, 22],
                 [175, 75, 28],
                 [160, 55, 35]])

print("Оригінальні дані:")
print(data)

# Обчислюємо середнє та стандартне відхилення по кожній ознаці
mean = data.mean(axis=0)  # shape (3,)
std = data.std(axis=0)    # shape (3,)

print(f"\nСереднє: {mean}")
print(f"Стандартне відхилення: {std}")

# Нормалізація: (data - mean) / std
# NumPy автоматично розширює mean та std (broadcasting)
normalized = (data - mean) / std

print("\nНормалізовані дані:")
print(normalized)

# Перевірка: середнє має бути ~0, std ~1
print(f"\nНове середнє: {normalized.mean(axis=0)}")
print(f"Нове std: {normalized.std(axis=0)}")
```

### Приклад 26. Множення матриці на вектор

```python
import numpy as np

# Матриця 2×3
A = np.array([[1, 2, 3],
              [4, 5, 6]])

# Вектор розмірності 3
x = np.array([1, 0, 2])

# Множення матриці на вектор
b = A @ x

print("Матриця A (2×3):")
print(A)
print(f"\nВектор x (3,):")
print(x)
print(f"\nРезультат A @ x (2,):")
print(b)
```

### Приклад 27. Обчислення лінійної комбінації ознак

```python
# Dataset: 4 студенти, 3 предмети
grades = np.array([[85, 90, 95],   # студент 0
                   [75, 80, 85],   # студент 1
                   [95, 95, 100],  # студент 2
                   [60, 70, 75]])  # студент 3

# Ваги важливості предметів
weights = np.array([0.3, 0.3, 0.4])

print("Оцінки студентів:")
print(grades)
print(f"\nВаги предметів: {weights}")

# Обчислюємо зважений бал для кожного студента
final_scores = grades @ weights

print(f"\nФінальні бали:")
for i, score in enumerate(final_scores):
    print(f"  Студент {i}: {score:.2f}")
```

### Приклад 28. Векторизація обчислень

```python
# 100 зразків, 5 ознак
X = np.random.randn(100, 5)

# Вектор ваг для 5 ознак
w = np.array([0.5, -0.3, 0.8, 0.2, -0.1])

# Обробка всіх 100 зразків одночасно
predictions = X @ w

print(f"Розмірність даних: {X.shape}")
print(f"Розмірність ваг: {w.shape}")
print(f"Розмірність результату: {predictions.shape}")
print(f"\nПерші 5 передбачень: {predictions[:5]}")
```

### Приклад 29. Множення матриць

```python
import numpy as np

# Матриці з прикладу
A = np.array([[1, 2, 3],
              [4, 5, 6]])  # 2×3

B = np.array([[7, 8],
              [9, 10],
              [11, 12]])   # 3×2

# Множення A @ B
C = A @ B

print("Матриця A (2×3):")
print(A)
print("\nМатриця B (3×2):")
print(B)
print(f"\nРезультат C = A @ B ({C.shape[0]}×{C.shape[1]}):")
print(C)
```

### Прииклад 30. Обчислення рангу матриці

```python
import numpy as np

# Матриця, у якій другий рядок — це подвоєний перший
A = np.array([
    [1, 2, 3],
    [2, 4, 6],
    [1, 1, 1]
])

print("Матриця A:")
print(A)
print(f"\nРозмір матриці: {A.shape[0]} × {A.shape[1]}")

rank = np.linalg.matrix_rank(A)
print(f"Ранг матриці: {rank}")
```
