Ви працюєте DevOps-інженером і аналізуєте логи системи моніторингу. За добу зафіксовано 500 подій. З них:

- 85 — помилки тільки бази даних з тегом DB
- 60 — помилки тільки мережі з тегом NET
- 35 — помилки, що мають обидва теги одночасно DB та NET

Позначимо:

$Ω$ — множина всіх 500 подій

$A$ — подія "помилка бази даних" з тегом DB

$B$ — подія "помилка мережі" з тегом NET


Завдання:

- Яка ймовірність, що випадково вибрана подія — помилка DB?

$A = (A \setminus B) + (A \cap B) = 85 + 35 = 120$

$P(A) = \frac{|A|}{|Ω|} = 120/500 = 12/50$ (24%)

- Яка ймовірність, що подія має тег DB або NET?

$(A \cup B) = (A \setminus B) + (B \setminus A) + (A \cap B) = 85 + 60 + 35 = 180$

$P(A \cup B) = \frac{|(A \cup B)|}{|Ω|} = 180/500 = 18/50$ (36%)

- Яка ймовірність, що подія має тег DB, але не NET?

$P(A \setminus B) = 85/500 $ (17%)

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

v = venn2(subsets = (85, 60, 35), set_labels = ('DB', 'NET'))

v.get_patch_by_id('10').set_color("#00DFF8")
v.get_patch_by_id("10").set_linewidth(3.0)
v.get_patch_by_id('01').set_color("#E733F8")
v.get_patch_by_id("01").set_linewidth(3.0)
v.get_patch_by_id('11').set_color("#00FB69")
v.get_patch_by_id("11").set_linewidth(3.0)

plt.title("Errors:")
plt.show()
```
