```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
import pandas as pd
```

```python

users = pd.read_csv('users.csv')
# print(users)    
#  Create a mask to filter dataset
active_mask = users["Is_Active"] == True
premium_mask = users["Is_Premium"] == True

# Create dataset by mask
act_users = users[active_mask]
p_users = users[premium_mask]

print(f"Active Users:\n {act_users}\n")
print(f"Premium Users:\n {p_users}")

act_users_set = set(act_users["User ID"])
p_users_set = set(p_users["User ID"])
print(f"Active Users Set: {act_users_set}")
print(f"Premium Users Set: {p_users_set}")

# Find users who are not active and premium
difference = p_users_set - act_users_set
print(f"Not Active and Premium: {difference}")


#  Plot
fig, ax = plt.subplots(1, 1, figsize=(8, 6))
v = venn2([act_users_set, p_users_set], set_labels=('Active', 'Premium'))
v.get_patch_by_id('10').set_color("#4CA511")
v.get_patch_by_id('11').set_color("#4CA511")
v.get_patch_by_id('01').set_color("#A51111")
```

```python
symmetric_diff = act_users_set.symmetric_difference(p_users_set)
print(symmetric_diff)

diff = act_users_set - p_users_set
print(diff)

#  Plot
fig, ax = plt.subplots(1, 1, figsize=(8, 6))
v = venn2([act_users_set, p_users_set], set_labels=('Active', 'Premium'))
v.get_patch_by_id('10').set_color("#8CDD11")
v.get_patch_by_id('11').set_color("#C3600A")
v.get_patch_by_id('01').set_color("#A51111")
```
